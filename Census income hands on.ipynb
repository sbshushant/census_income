### Linear regression
a. Build a linear regression model as follows:
1) Divide the dataset into training and test sets in 70:30 ratio
2) Build a linear model on the test set where the dependent variable is 'hours-per-week' and the independent variable is 'education-num'
3) Predict the values on the train set and find the error in prediction 
4) Find the root-mean-swuare error (RMSE)

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X = df[['education-num']]
Y = df['hours-per-week']            #y=df.loc[:,['hours-per-week']]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,train_size=.3, random_state=0)
```

```python
l1 = LinearRegression()
```

```python
l1.fit(X_train,Y_train)
```

```python
pred = l1.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
rmse = np.sqrt(mean_squared_error(Y_test,pred))
```

```python
rmse
```

### Logistic regression
a) Build a simple logistic regression model as follows:
1) Divide the dataset into training and test set in 65:35 ratio 
2) Build a logistic regression model where the dependent variable is 'x' (yearly income) and the independent variable is 'occupation'
3) Predict the values on the test set 
4) Build the confusion matrix and find the accuracy

```python
from sklearn.preprocessing import LabelEncoder     #use to convert object into numerical column
```

```python
le = LabelEncoder()
```

```python
for i in df.columns:
    if df[i].dtype == 'object':
        df[i]= le.fit_transform(df[i])
```

```python
df
```

```python
X = df[['occupation']]
Y = df['AnnualIncome']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=.35, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(X_train,Y_train)
```

```python
pred1 = lr.predict(X_test)
```

```python
pred1
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
accuracy_score(Y_test,pred1)
```

```python
confusion_matrix(Y_test,pred1)
```
